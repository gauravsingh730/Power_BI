###**Question 1 :**
###Define Power BI and What are the key components of the Power BI ecosystem? Brieflyexplain:

- Power BI Desktop

- Power BI Service

- Power BI Mobile

- Power BI Gateway

**Answer :**

Power BI is a business intelligence (BI) and data visualization tool developed by Microsoft that enables users to connect to data from multiple sources, transform and model the data, and create interactive reports and dashboards for data-driven decision-making.

**Key Components of the Power BI Ecosystem**

**1. Power BI Desktop**

- A Windows-based application used for data preparation, modeling, and report creation.

- Allows users to connect to various data sources, clean and transform data using Power Query, and build data models using DAX (Data Analysis Expressions).

- Reports created in Power BI Desktop are published to the Power BI Service for sharing.

**2. Power BI Service**

- A cloud-based platform (SaaS) used to publish, share, and collaborate on reports and dashboards.

- Enables users to create dashboards, set up data refresh schedules, and control access and security.

- Supports collaboration through workspaces and sharing with others in the organization.

**3. Power BI Mobile**

- Mobile applications available for Android, iOS, and Windows devices.

- Allows users to view and interact with reports and dashboards anytime, anywhere.

- Provides real-time alerts and touch-optimized visualizations for mobile use.

**4. Power BI Gateway**

- Acts as a bridge between on-premises data sources and the Power BI Service.

- Enables secure data transfer for scheduled refresh and live queries without moving data to the cloud.

- Commonly used to connect Power BI Service with on-premises databases like SQL Server or Oracle.

In summary, **Power BI Desktop** is used for building reports, the Power BI Service is for sharing and collaboration, Power BI Mobile provides access on the go, and the Power BI Gateway ensures secure connectivity to on-premises data sources.

###**Question 2 :** Compare the following Power BI visuals:

- Pie Chart vs Donut Chart

- Bar Chart vs Column Chart

When would you prefer one over the other? Give one example for each pair.

**Answer:**

**1. Pie Chart vs Donut Chart**

| Aspect        | Pie Chart                                   | Donut Chart                                   |
| ------------- | ------------------------------------------- | --------------------------------------------- |
| Structure     | Circular chart divided into slices          | Similar to pie chart but with a hollow center |
| Data clarity  | Better for simple part-to-whole comparisons | Allows space in the center for labels or KPIs |
| Visual appeal | Simple and traditional                      | More modern and visually appealing            |

**When to prefer which**

- Pie Chart: Use when you have few categories (2–4) and want a simple part-to-whole comparison.

- Donut Chart: Use when you want to display an additional value (like total) in the center or improve visual aesthetics.

**Example**

- Pie Chart: Market share of mobile operating systems (Android, iOS, Others).

- Donut Chart: Percentage of sales by region with total sales shown in the center.

**2. Bar Chart vs Column Chart**

| Aspect      | Bar Chart                               | Column Chart                         |
| ----------- | --------------------------------------- | ------------------------------------ |
| Orientation | Horizontal bars                         | Vertical columns                     |
| Best for    | Long category names and many categories | Time-based or sequential comparisons |
| Readability | Easier to read labels                   | Better for showing trends over time  |

**When to prefer which**

- Bar Chart: Use when category labels are long or when there are many categories.

- Column Chart: Use when comparing values across time periods or ordered data.

**Example**

- Bar Chart: Sales by product category with long names (e.g., “Home Appliances,” “Electronics Accessories”).

- Column Chart: Monthly sales comparison for a year.

**Summary:**

Pie vs Donut charts are best for part-to-whole analysis, with Donut charts offering better space utilization. Bar vs Column charts differ mainly in orientation, with Bar charts suited for categorical comparison and Column charts ideal for time-based data.

###**Question 3 :** Explain the significance of:

- Star schema vs Snowflake schema

- Primary key vs Foreign key in relationships (Power BI)

Why is cardinality important?

**Answer :**

**1. Star Schema vs Snowflake Schema**

**Star Schema**

- A data modeling design where a central fact table is directly connected to multiple dimension tables.

- Structure is simple and easy to understand.

- Improves query performance and is recommended for Power BI models.

**Snowflake Schema**

- A more complex design where dimension tables are normalized into multiple related tables.

- Reduces data redundancy but increases model complexity.

- Can slightly reduce performance due to additional joins.

**Significance / Comparison**

 | Aspect         | Star Schema        | Snowflake Schema                    |
| -------------- | ------------------ | ----------------------------------- |
| Complexity     | Simple             | Complex                             |
| Performance    | Faster             | Slower                              |
| Maintenance    | Easy               | Difficult                           |
| Power BI usage | Highly recommended | Used when normalization is required |


**2. Primary Key vs Foreign Key (in Power BI Relationships)**

**Primary Key**

- A column that uniquely identifies each record in a table.

- Typically found in dimension tables.

- Contains unique and non-null values.

**Foreign Key**

- A column that references the primary key of another table.

- Typically found in fact tables.

- Used to establish relationships between tables.

**Example**

- CustomerID in Customer table → Primary Key

- CustomerID in Sales table → Foreign Key

**Why Is Cardinality Important?**

Cardinality defines the relationship type between tables in Power BI:

- One-to-Many (1:*) – Most common and recommended

- Many-to-One (*:1) – Reverse of one-to-many

- One-to-One (1:1)

- Many-to-Many (:)

**Significance of Cardinality**

- Ensures accurate data filtering and aggregation.

- Prevents duplicate values and incorrect totals.

- Improves model performance and DAX calculation accuracy.

- Helps Power BI understand how data flows between tables.

**Example**

- One Customer can have many Sales → 1: relationship*



###**Question 4 :** Differentiate between:
- Calculated column vs Measure

Also, define Row context and Filter context with simple examples.

**Answer :**

| Aspect          | Calculated Column       | Measure                           |
| --------------- | ----------------------- | --------------------------------- |
| When calculated | During **data refresh** | Calculated **at query time**      |
| Storage         | Stored in the model     | Not stored (calculated on demand) |
| Context         | Uses **row context**    | Uses **filter context**           |
| Use case        | Row-level calculations  | Aggregations and KPIs             |
| Performance     | Increases model size    | More efficient for aggregations   |

**Example**

- Calculated Column:

Profit = Sales[Revenue] - Sales[Cost]

(Calculated for each row and stored)

- Measure:

Total Sales = SUM(Sales[Revenue])

(Calculated dynamically based on filters)

**2. Row Context**

**Definition:**

Row context means calculations are evaluated row by row in a table.

**Where it occurs:**

- Calculated columns

- Iterators like SUMX, AVERAGEX

**Simple Example**

- In a Sales table:

Profit = Sales[Revenue] - Sales[Cost]

Each row calculates profit using that row’s revenue and cost.

**3. Filter Context**

**Definition:**

Filter context is the set of filters applied to data (from slicers, visuals, or report filters) that affect a calculation.

**Where it occurs:**

- Measures

- Visuals, slicers, and page/report filters

**Simple Example**

- Measure:
Total Sales = SUM(Sales[Revenue])

- If a slicer filters Year = 2025, the measure returns sales for 2025 only.

**Key Difference Between Row Context and Filter Context**

| Aspect   | Row Context        | Filter Context             |
| -------- | ------------------ | -------------------------- |
| Scope    | Single row         | Subset of data             |
| Used in  | Calculated columns | Measures                   |
| Dynamic? | No (static)        | Yes (changes with filters) |

###**Question 5: What is the difference between a report and a dashboard in Power BI?**

**Answer :**

| Aspect        | **Report**                                                         | **Dashboard**                                                                |
| ------------- | ------------------------------------------------------------------ | ---------------------------------------------------------------------------- |
| Definition    | A **collection of pages** with visuals built from a single dataset | A **single-page canvas** made of pinned visuals from one or multiple reports |
| Data source   | Uses **one dataset**                                               | Can combine visuals from **multiple datasets**                               |
| Interactivity | Highly interactive (filters, slicers, drill-through)               | Limited interactivity                                                        |
| Creation tool | Created in **Power BI Desktop**                                    | Created in **Power BI Service**                                              |
| Purpose       | Detailed analysis and exploration                                  | High-level summary and monitoring                                            |
| Drill-down    | Supported                                                          | Not supported                                                                |

**When to Use Each**

- **Report:**
  - Use when you need detailed analysis, multiple views of data, and interactive exploration.
Example: A sales performance report with pages for region-wise, product-wise, and time-based analysis.

- **Dashboard:**
  - Use when you want a quick overview of key metrics on a single screen.
Example: An executive dashboard showing total revenue, profit, and top-performing
